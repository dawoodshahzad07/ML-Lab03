```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\Home\\Desktop\\Automobile_data.csv')
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 50)
df.head(100)
```

```python
df.isnull().sum()
```

```python
df['normalized-losses']=df['normalized-losses'].replace({'?':None})
```

```python
df['normalized-losses'].unique()
```

```python
df['normalized-losses']=pd.to_numeric(df['normalized-losses']).fillna(0).astype('int')
```

```python
df['normalized-losses'].unique()
```

```python
losses_mean=df['normalized-losses'].mean()
losses_mean
```

```python
df['normalized-losses']=df['normalized-losses'].replace({0:losses_mean})
```

```python
col=df.columns

for col in df:
    print("\n-------->",col,"<---------")
    print(df[col].unique(),df[col].dtype)
   
```

```python
columns=df.columns

for columns in df:
    if '?' in df[columns].unique():
        df[columns]=df[columns].replace({'?':0})
```

```python
from pandas import to_numeric
df[['stroke','horsepower','peak-rpm','price']]=df[['stroke','horsepower','peak-rpm','price']].apply(pd.to_numeric,errors='coerce')
col=['stroke','horsepower','peak-rpm','price']
df.dtypes

for col_name in col:
    col_mean = df[col_name].mean()
    df[col_name] = df[col_name].replace({0: col_mean})
    
```

```python

df['num-of-doors']=df['num-of-doors'].replace({'two':2})
df['num-of-doors']=df['num-of-doors'].replace({'four':4})
df['num-of-doors']=pd.to_numeric(df['num-of-doors']).astype(int)
```

```python
df['num-of-doors']=df['num-of-doors'].drop(index=27)
df['num-of-doors']=df['num-of-doors'].drop(index=63)
df[df['num-of-doors']==0]
```

```python

df['num-of-cylinders']=df['num-of-cylinders'].replace({'two':2})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'three':3})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'four':4})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'five':5})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'six':6})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'eight':8})
df['num-of-cylinders']=df['num-of-cylinders'].replace({'twelve':12})
df['num-of-cylinders']=pd.to_numeric(df['num-of-cylinders']).astype(int)
df['num-of-cylinders'].unique()
```

```python

df['bore']=pd.to_numeric(df['bore']).astype(int)
df['bore']=df['bore'].replace({0:1})
df['bore'].unique()
```
